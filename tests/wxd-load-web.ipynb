```python
import requests
from bs4 import BeautifulSoup

url = "https://www.ibm.com/products/watsonx-data"  # Replace with the URL of the web page you want to retrieve
response = requests.get(url)
content = response.text

soup = BeautifulSoup(content, 'html.parser')
text = soup.get_text()

print(text)
```
